```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 这里使用的是黑体作为例子
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 读取 CSV 文件
file_path = 'predictions.csv'  # 替换成你的 CSV 文件路径
data = pd.read_csv(file_path)

# 统计每个分类的数量
category_counts = data.iloc[:, 1].value_counts().sort_index()  # 第二列的分类统计
plt.figure(figsize=(10, 4))
# 绘制柱状图
plt.bar(["国家战略支撑股","关键技术龙头股","核心基建重工股","长期发展扶持股","科技创新独角兽股","其他多元概念股"], category_counts.values)
plt.xlabel('类别')
plt.ylabel('数量')
plt.title('部分沪深A股分类结果')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 损失值列表示例（请替换为您的实际值）
loss_values = [0.5, 0.4, 0.3, 0.2, 0.15, 0.1, 0.08, 0.06, 0.005]

# 生成平滑曲线
smoothed_loss_values = np.convolve(loss_values, np.ones(3)/3, mode='valid')

# 创建一个图表来显示平滑曲线
plt.figure(figsize=(8, 6))
plt.plot(smoothed_loss_values, linestyle='-', color='blue')
plt.title('损失随Epoch变化（平滑曲线）')
plt.xlabel('Epoch')
plt.ylabel('损失')
plt.grid(True)
plt.show()
```

```python
 from visdom import Visdom
 from sklearn.datasets import  load_iris
 import torch
 import numpy as np
 from PIL import Image
 iris_x, iris_y = load_iris(return_X_y=True)

 # 获取绘图对象，相当于plt
 vis = Visdom()
 # 添加折线图
 x = torch.linspace(-6, 6, 100).view([-1, 1])
 sigmoid = torch.nn.Sigmoid()
 sigmoid_y = sigmoid(x)
 tanh = torch.nn.Tanh()
 tanh_y = tanh(x)
 relu = torch.nn.ReLU()
 relu_y = relu(x)
 # 连接三个张量
 plot_x = torch.cat([x, x, x], dim=1)
 plot_y = torch.cat([sigmoid_y, tanh_y, relu_y], dim=1)
 # 绘制线性图
 vis.line(X=plot_x, Y=plot_y, win="line plot", env="main",
          opts={
              "dash" : np.array(["solid", "dash", "dashdot"]),
              "legend" : ["Sigmoid", "Tanh", "ReLU"]
          })
```
