```python
import os  
import pandas as pd  
   
folder1 = '合并'  
folder2 = '总的价值投资指标'  
 
 
output_folder = '价值指标的最终合并成品'  
  

security_codes = os.listdir(folder1)  
for security_code in security_codes:    
    # 初始化一个标志变量，用于记录三个文件是否都存在  
    all_files_exist = True  
  
    # 检查文件是否在三个文件夹中都存在    
    file1_path = os.path.join(folder1, security_code)    
    file2_path = os.path.join(folder2, security_code)       
  
    # 检查三个文件是否存在，并且是否是Excel文件  
    if not (os.path.isfile(file1_path) and os.path.isfile(file2_path)):    
        all_files_exist = False  
    else:  
        # 如果是Excel文件（这里假设所有文件都是.xlsx格式）    
        if not (file1_path.endswith('.xlsx') and file2_path.endswith('.xlsx') ):    
            all_files_exist = False  
  
    # 如果所有文件都存在，才执行后续的操作  
    if all_files_exist:  
        # 从三个Excel文件中抽取指定列    
        # 从第一个表格中抽取指定列  
        file1_columns = ["股票名称","总市值","流通市值","所处行业","市盈率(动)","市净率","ROE","毛利率","净利率","板块编号"]  
        file1 = pd.read_excel(file1_path, usecols=file1_columns)   
        # 从第二个表格中抽取指定列  
        file2_columns = ["年份", "季度", "证券代码", "公司发布财报的日期", "财报统计的季度的最后一天", "流动比率", "速动比率", "现金比率", "总负债同比增长率", "资产负债率", "权益乘数","流动资产除以总资产", "有形资产除以总资产", "已获利息倍数", "经营活动产生的现金流量净额除以营业收入", "经营性现金净流量除以净利润", "	经营性现金净流量除以营业总收入","净资产收益率(平均)(%)", "销售净利率(%)", "销售毛利率(%)", "净利润(元)", "每股收益", "主营营业收入(元)", "总股本", "流通股本"]  
        file2 = pd.read_excel(file2_path, usecols=file2_columns)       
        file1 = pd.read_excel(file1_path, usecols=file1_columns)     
        file2 = pd.read_excel(file2_path, usecols=file2_columns)   
                         
              
        # 使用pd.concat()函数将三个DataFrame横向合并    
        merged_df = pd.concat([file2, file1], axis=1)    
                
        # 将合并后的DataFrame保存到新的Excel文件中，并存放到指定的文件夹    
        output_path = os.path.join(output_folder, f"{security_code}")    
        merged_df.to_excel(output_path, index=False)  
    else:  
        print(f"One or more files for {security_code} do not exist. Skipping...")  
        continue
```

```python
import os  
import re  
  
# 设置文件夹路径  
folder_path = '每个行业前面20个'  
  
# 遍历文件夹中的文件  
for filename in os.listdir(folder_path):  
    # 检查文件是否为.xlsx文件  
    if filename.endswith('.xlsx'):  
        # 使用正则表达式匹配中文字符，并生成新文件名  
        new_filename = re.sub(r'[^\u4e00-\u9fa5]', '', filename)  
  
        # 添加.xlsx后缀  
        new_filename = new_filename + '.xlsx'  
  
        # 如果新文件名与原始文件名不同，并且新文件名不为空，则进行重命名操作  
        if new_filename != filename and new_filename:  
            # 生成文件的完整路径  
            file_path = os.path.join(folder_path, filename)  
            new_file_path = os.path.join(folder_path, new_filename)  
  
            # 重命名文件，覆盖原文件  
            os.rename(file_path, new_file_path)
```
