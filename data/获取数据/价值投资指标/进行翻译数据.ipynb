```python
import os  
import pandas as pd  
   
folder1 = '国企沪市的季频偿债能力'  
folder2 = '国企沪市的季频现金流数据'  
folder3 = '国企沪市的季频盈利能力'  
  

output_folder = '国企沪市的三合一'  
  

security_codes = os.listdir(folder1)  
for security_code in security_codes:    
    # 初始化一个标志变量，用于记录三个文件是否都存在  
    all_files_exist = True  
  
    # 检查文件是否在三个文件夹中都存在    
    file1_path = os.path.join(folder1, security_code)    
    file2_path = os.path.join(folder2, security_code)    
    file3_path = os.path.join(folder3, security_code)    
  
    # 检查三个文件是否存在，并且是否是Excel文件  
    if not (os.path.isfile(file1_path) and os.path.isfile(file2_path) and os.path.isfile(file3_path)):    
        all_files_exist = False  
    else:  
        # 如果是Excel文件（这里假设所有文件都是.xlsx格式）    
        if not (file1_path.endswith('.xlsx') and file2_path.endswith('.xlsx') and file3_path.endswith('.xlsx')):    
            all_files_exist = False  
  
    # 如果所有文件都存在，才执行后续的操作  
    if all_files_exist:  
        # 从三个Excel文件中抽取指定列    
        # 从第一个表格中抽取指定列
         
        file1_columns = ["年份", "季度", "证券代码", "公司发布财报的日期", "财报统计的季度的最后一天", "流动比率", "速动比率", "现金比率", "总负债同比增长率", "资产负债率", "权益乘数"]  
        try: 
            file1 = pd.read_excel(file1_path, usecols=file1_columns)  
        except:
            print(f"{file1_path}缺少一些列，从而跳过处理...") 
            continue
        # 从第二个表格中抽取指定列  
        
        file2_columns = ["流动资产除以总资产", "有形资产除以总资产", "已获利息倍数", "经营活动产生的现金流量净额除以营业收入", "经营性现金净流量除以净利润", "	经营性现金净流量除以营业总收入"]  
        try: 
            file2 = pd.read_excel(file2_path, usecols=file2_columns) 
        except:
            print(f"{file2_path}缺少一些列，从而跳过处理...") 
            continue
        # 从第三个表格中抽取指定列  
        file3_columns = ["净资产收益率(平均)(%)", "销售净利率(%)", "销售毛利率(%)", "净利润(元)", "每股收益", "主营营业收入(元)", "总股本", "流通股本"]  
        try:
            file3 = pd.read_excel(file3_path, usecols=file3_columns) 
        except: 
            print(f"{file3_path}缺少一些列，从而跳过处理...") 
            continue      
        
        # file1 = pd.read_excel(file1_path, usecols=file1_columns)     
        # file2 = pd.read_excel(file2_path, usecols=file2_columns)   
        # file3 = pd.read_excel(file3_path, usecols=file3_columns)                            
              
        # 使用pd.concat()函数将三个DataFrame横向合并    
        merged_df = pd.concat([file1, file2, file3], axis=1)    
                
        # 将合并后的DataFrame保存到新的Excel文件中，并存放到指定的文件夹    
        output_path = os.path.join(output_folder, f"{security_code}")    
        merged_df.to_excel(output_path, index=False)  
    else:  
        print(f"One or more files for {security_code} do not exist. Skipping...")  
        continue
```

```python
import os  
import pandas as pd  
  
# 设置两个文件夹的路径  
folder_path_1 = '国有每个行业抽10个'  
folder_path_2 = '非国有每个行业抽10个'  
  
# 设置保存合并后的Excel文件的空文件夹路径  
save_folder_path = '前面20个'  # 将这个路径改为您的空文件夹路径  
  
# 确保保存文件夹存在，如果不存在则创建  
if not os.path.exists(save_folder_path):  
    os.makedirs(save_folder_path)  
  
# 遍历文件夹1中的Excel文件  
for filename in os.listdir(folder_path_1):  
    if filename.endswith(".xlsx") or filename.endswith(".xls"):  
        file_path_1 = os.path.join(folder_path_1, filename)  
        file_path_2 = os.path.join(folder_path_2, filename)  
        base_filename, _ = os.path.splitext(filename)  # 获取文件名不包括扩展名的部分  
  
        # 读取文件夹1和文件夹2中的同名Excel文件  
        df1 = pd.read_excel(file_path_1)  
        df2 = pd.read_excel(file_path_2)  
  
        # 使用concat方法进行数据合并，按列标签进行合并，默认是外连接（outer join）  
        merged_df = pd.concat([df1, df2], axis=0, ignore_index=True)  
          
        # 将合并后的数据保存到指定的空文件夹中，并使用原始文件名保存  
        save_file_path = os.path.join(save_folder_path, base_filename + '_merged.xlsx')  
        merged_df.to_excel(save_file_path, index=False)
```
