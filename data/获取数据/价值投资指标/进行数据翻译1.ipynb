```python
import os  
import pandas as pd  
  
# 设置原始文件夹和新的目标文件夹  
source_folders = ['非国企沪市的季频偿债能力', '非国企沪市的季频现金流数据', '非国企沪市的季频盈利能力']  
target_folder = '非国企沪市的三合一'  
  
# 确保目标文件夹存在  
if not os.path.exists(target_folder):  
    os.makedirs(target_folder)  
  
# 定义共同的列名  
common_columns = ['证券代码', '公司发布财报的日期', '财报统计的季度的最后一天', '年份', '季度']  
  
merged_df = pd.DataFrame()  # 初始化一个空的DataFrame  
  
# 定义一个函数来为冲突的列名添加后缀  
def add_suffix_to_columns(df, suffix):  
    new_columns = {}  
    for col in df.columns:  
        new_columns[col] = col + suffix  
    df.columns = new_columns  
    return df  
  
# 遍历每个原始文件夹  
for folder in source_folders:  
    for filename in os.listdir(folder):  
        if filename.endswith('.xlsx'):  
            filepath = os.path.join(folder, filename)  
            df = pd.read_excel(filepath)  
              
            # 检查当前文件中的列名是否与已合并的DataFrame中的列名冲突  
            conflicting_columns = set(df.columns) & set(merged_df.columns) - set(common_columns)  
              
            # 如果有冲突，为冲突的列名添加文件名的后缀  
            if conflicting_columns:  
                suffix = '_' + os.path.splitext(os.path.basename(filepath))[0]  
                df = add_suffix_to_columns(df, suffix)  
              
            # 只保留共同的列和其他独有的列  
            df = df[common_columns + [col for col in df.columns if col not in common_columns and col not in merged_df.columns]]  
              
            # 使用outer merge合并数据  
            if merged_df.empty:  
                merged_df = df  
            else:  
                merged_df = pd.merge(merged_df, df, on=common_columns, how='outer')  
            # 将当前的文件名作为目标文件名  
            target_file = filename  # 更新目标文件名为当前文件名  

# 将合并后的数据保存到目标文件夹中，使用指定的文件名  
merged_df.to_excel(os.path.join(target_folder, target_file), index=False)  
  
print("合并完成！")
```

```python
import os  
import pandas as pd  
  
# 读取原始Excel文件  
original_file = '非国有价值投资指标合并数据.xlsx'  # 替换为你的原始Excel文件路径  
df = pd.read_excel(original_file)  
  
# 指定要保存新Excel文件的文件夹路径  
save_folder = '非国有每个行业抽10个'  # 替换为你要保存新Excel文件的文件夹路径  
  
# 筛选每个行业的股票，并抽取10个数据  
unique_industries = df['所处行业'].unique()  
for industry in unique_industries:  
    # 筛选指定行业的股票数据  
    industry_data = df[df['所处行业'] == industry].head(10)  # 使用head(10)抽取前10个数据  
  
    # 生成新的Excel文件名并保存到指定文件夹  
    new_file_name = f'{industry}_top10.xlsx'  
    new_file_path = os.path.join(save_folder, new_file_name)  
  
    # 将筛选后的数据保存到新的Excel文件  
    industry_data.to_excel(new_file_path, index=False)
```
